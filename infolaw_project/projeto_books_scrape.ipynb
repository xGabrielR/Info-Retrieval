# Agora fazer para todas as paginas

```python
# Definindo o número maximo de paginas do site:
max_pag = soup.find('li', class_='current').get_text(strip=True)
max_pag = re.sub('\D', '', max_pag)
max_pag = int(max_pag[1:]) # Porque esse link sempre estara na pagina 1, então pega do caractere 1 em diante. Ignora o 0
max_pag
```

```python
# Criando o DataFrame que ira ser preenchido com os dados:
data_final = pd.DataFrame()

# Iterando sobre todas as paginas para coletar TODOS os produtos:
for i in range(1, max_pag + 1):
    # Definindo as url's:
    url = f'https://books.toscrape.com/catalogue/page-{i}.html'
    
    # Fazendo o request:
    page = requests.get(url)

    # ------- Instanciamento do BS4 -------
    soup = BeautifulSoup(page.text, 'html.parser')
    
    # ------- Criando as variaveis que irei trabalhar -------
    product_name = []
    product_price = []
    product_avaible = []
    product_rating = []
    product_link = []

    # ------- -------
    # Definindo a vitrine de produtos para buscar em cima dela:
    product_list = soup.find_all('article', class_='product_pod')

    # ------- Product name -------
    for products in product_list:
        product_name.append(products.find('img', class_='thumbnail').get('alt'))    

    # ------- Product Price -------
    for products in product_list:
        product_price.append(products.find('p', class_='price_color').get_text())    

    # -------Product avaible-------
    for products in product_list:
        product_avaible.append(products.find('p', class_='instock availability').get_text(strip=True))

    # ------- product_rating -------
    rating_list = soup.find_all('p', class_=re.compile('^star-rating'))
    for ratings in rating_list:
        product_rating.append(ratings.get('class'))

    # ------- product_link -------
    for product in product_list:
        product_link.append('https://books.toscrape.com/' + product.find('div', class_='image_container').find('a').get('href'))

    # ------- Criando o DataFrame e nomeando suas colunas -------
    data = pd.DataFrame([product_name, product_price, product_avaible, product_rating, product_link]).T
    data.columns = ['product_name', 'product_price', 'product_avaible', 'product_rating(stars)', 'product_link']


    # ------- Limpando e formatando a coluna preço (Deletando qualquer caractere nao numerico e repondo o ponto) -------
    data['product_price'] = data['product_price'].apply(lambda x: re.sub('\D', '', x))
    data['product_price'] = data['product_price'].apply(lambda x: ''.join((x[:2], '.', x[3:])))

    # ------- # Alterando o conteudo das avaliações para númerico (1 a 5) --------

    data['product_rating(stars)'] = data['product_rating(stars)'].apply(lambda x: 1 if x[1] == 'One'
                                                            else 2 if x[1] == 'Two'
                                                            else 3 if x[1] == 'Three'
                                                            else 4 if x[1] == 'Four'
                                                            else 5 if x[1] == 'Five'
                                                            else 'NA')


    data_final = pd.concat([data_final, data], ignore_index=True)


data_final.head(50)
```
