```python
import pandas as pd

from elasticsearch import Elasticsearch
from sqlalchemy import create_engine
```

# Conexão com Nuvem

```python
#Conexao sql
db = create_engine()
con = db.connect()

#Conexao Elasticsearch
es = Elasticsearch(
    cloud_id=,
    basic_auth=()
)
```

# Match

```python
#match Searching per only one character
es.search(index="airbnb_houses", body={
  "query": {
    "match": {
      "neighbourhood_group": "Manhattan"
    }
  }
})

#Match searching several characters
es.search(index="airbnb_houses", body={"query": 
                                          { "match": 
                                             {"name":'clean | of | junior '
                                               }
                                          }
                                      }).body


#sql 
pd.read_sql_query("SELECT * FROM airbnb_houses WHERE neighbourhood_group IN ('Manhattan', 'of', 'junior')", con=con)
```

# Match_all

```python
#match_all
es.search(index="airbnb_houses", body={"query": { "match_all": {}
                                                }
                                      })
#Sql 
pd.read_sql_query("select * from airbnb_houses", con=con)
```

# Range 

```python
es.search(index="airbnb_houses", body={
  "query": {
    "range": {
      "price": {
        "gte": 79,
        "lte": 325
      
      }
    }
  }
    }).body
#Sql
pd.read_sql_query("SELECT * FROM airbnb_houses WHERE (price >= 79) AND (price <= 325)", con=con)
```

# Prefix

```python
#prefix 
es.search(index="airbnb_houses", body={
  "query": {
    "prefix": {
      "host_name": {
        "value": "La"
      }
    }
  }
}).body     


#Sql 
pd.read_sql_query("select * FROM airbnb_houses WHERE host_name LIKE 'La%%' ", con=con)
```

# Term

```python
#Term
es.search(index="airbnb_houses", body={
  "query": {
    "term": {   
      "host_id": {
        "value":"9744",
        "boost": 1.7,
        "case_insensitive":False  
      }
    }       
  }
}).body

#Sql
pd.read_sql_query("SELECT * FROM airbnb_houses WHERE host_id = 9744 ", con=con)
```

# Boolean

```python
#Must
es.search(index="airbnb_houses", body={
  "query": {
    "bool" : {
      "must" : {
        "range" : { 
            "host_id" : {'gte':'320538',
                         'lte':'1191142'}
        }
      }
    }
  }
}).body



#Sql
pd.read_sql_query("SELECT * FROM airbnb_houses WHERE (host_id >= 320538) AND (host_id <= 1191142)", con=con)
```

```python
# Filter
es.search(index="airbnb_houses", body={
    "query": {
      "bool" : {
        "filter": {
          "term" : {"room_type": "Private room"}
      },"boost" : 1.0}}}).body

#Sql
pd.read_sql_query("SELECT * FROM airbnb_houses WHERE room_type = 'Private room'", con=con)
```

```python
# Must_not
es.search(index="airbnb_houses", body={
 "query": {
    "bool" : {
      "must_not" : {
         "range" : {
            "number_of_reviews" : { "lte" : 155, "gte" : 9 }
     }}}}}).body

#Sql
pd.read_sql_query("SELECT * FROM airbnb_houses WHERE number_of_reviews NOT BETWEEN 9 AND 155", con=con)
```

```python
# Should
es.search(index="airbnb_houses", body= {
        "query": {
           "bool" : {
               "should" : [
                 { "term" : { "neighbourhood" : "Williamsburg" } },
                 { "term" : { "host_name" : "Larissa" } }],
                  "minimum_should_match" : 1,
                  "boost" : 1.0
           }}}).body

#Sql
pd.read_sql_query("SELECT distinct * FROM airbnb_houses WHERE (neighbourhood = 'Williamsburg' OR host_name = 'Larissa')", con=con)
```

# Fuzzy

```python
#Fuzzy
es.search(index="airbnb_houses",size=10, body={
  "query": {
    "fuzzy": {
      "host_name": {
        "value": "aya"
      }  
    }
  }
}).body

#Sql
pd.read_sql_query("SELECT * FROM airbnb_houses WHERE host_name LIKE '%%Maya' ", con=con)
```
